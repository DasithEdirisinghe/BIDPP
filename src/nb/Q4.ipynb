```python
import scipy.stats as stats

# F1 scores of the two models
f1_model1 = 0.8956961251680722
f1_model2 = 0.8823469013568023

# Perform a two-sample t-test
t_stat, p_value = stats.ttest_ind([f1_model1], [f1_model2], equal_var=True)

# Print the results
print("T-statistic:", t_stat)
print("P-value:", p_value)
```
